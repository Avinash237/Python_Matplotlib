```python
# plot tempreture data
temp = [27,29,30,19,20]
```

Line plot

```python
plt.plot(temp)  # index of each number in x axis and values are in y axis
```

```python
plt.plot(['day1','day2','day3','day4','day51'],temp)
```

```python
# change color
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta')
```

```python
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--')
```

```python
# only markers will print
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle=' ',marker='o')
```

```python
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('5 Days temp data')
```

```python
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o',label='Daywise temp.')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('5 Days temp data')
plt.legend()
```

# add 2 line plots in one graph

```python
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o',label='Jan-2022')
plt.plot(['day1','day2','day3','day4','day51'],[18,20,35,33,19],color='green',linestyle='-',marker='*',label='Dec-2022')
plt.legend()
plt.savefig('temp.png')   # to save the image
```

```python
# if we want good resolution of saved image then use 'dpi' dot per inch
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o',label='Jan-2022')
plt.plot(['day1','day2','day3','day4','day51'],[18,20,35,33,19],color='green',linestyle='-',marker='*',label='Dec-2022')
plt.legend()
plt.savefig('temp.png',dpi=300)   # to save the image
```

```python
# if we want greed lines on graph use grid
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o',label='Jan-2022')
plt.plot(['day1','day2','day3','day4','day51'],[18,20,35,33,19],color='green',linestyle='-',marker='*',label='Dec-2022')
plt.legend()
plt.grid()
plt.savefig('temp.png',dpi=300)   # to save the image
```

```python
# increse or decrease the size
plt.figure(figsize=(4,4)) # provide width and height
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o',label='Jan-2022')
```

```python
plt.figure(figsize=(18,10)) # provide width and height
plt.plot(['day1','day2','day3','day4','day51'],temp,color='magenta',linestyle='--',marker='o',label='Jan-2022')
```

## Bar Plot

```python
# catagory vs count plot
company = ['infy','IBM','wipro','TCS']
share = [240,560,340,780]
```

```python
# catagoty is x axis and count is y axis
plt.bar(company,share,color='red')
```

```python
plt.bar(company,share,color=['red','green','blue','cyan'])
```

```python
# change x label
plt.bar(company,share,color=['red','green','blue','cyan'])
plt.xticks(range(4),['Infosys','IBM Tech','Wipro Tech','Tata Tech'])
```

```python
# skipp extra information  use plt.show
plt.bar(company,share,color=['red','green','blue','cyan'])
plt.xticks(range(4),['Infosys','IBM Tech','Wipro Tech','Tata Tech'])
plt.show()
```

```python
# rotation
plt.bar(company,share,color=['red','green','blue','cyan'])
plt.xticks(range(4),['Infosys','IBM Tech','Wipro Tech','Tata Tech'],rotation=45)
plt.show()
```

# Horizontal Bar plot: Barh

```python
# barh
plt.barh(company,share,color=['red','green','blue','cyan'])
plt.xticks(range(4),['Infosys','IBM Tech','Wipro Tech','Tata Tech'],rotation=45)
plt.show()
```

```python
# yticks  ## height
plt.barh(company,share,color=['red','green','blue','cyan'],height=.4)
plt.yticks(range(4),['Infosys','IBM Tech','Wipro Tech','Tata Tech'],rotation=45)
plt.show()
```

## Hist Plot

```python
# hist plot is a univarient plot, it display range wise distribution
# generate data of marks for 100 student
per = np.random.randint(42,100,100)
# (42 = min %,100 max %,100 total no. of student)
per
```

```python
plt.hist(per)
```

```python
# supply own range
plt.hist(per,bins=[42,50,65,85,99])
```

```python
# display the percentage of each year
plt.pie(share,autopct='%f%%')
plt.show()
```

```python
# make it 2 deciman only
plt.pie(share,autopct='%.2f%%')
plt.show()
```

```python
# seprate the pie
plt.pie(share,autopct='%.2f%%',labels=company,explode=[0,0,0,0.2])    # explode tcs .2 percent
plt.show()
```
